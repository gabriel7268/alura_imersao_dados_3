```python
# Importar Bibliotecas:
import pandas as pd
import matplotlib.pyplot as plt

# Link para Download da base de dados:
link = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

# Leitura da base de dados, arquivo compactado .zip:
dados = pd.read_csv(link, compression = 'zip')

# Imprime os dados, formato DataFrame:
display(dados)
## Cada linha do DataFrame é um experimento, cada coluna é um tipo de dado, parâmetro. Provavelmente teste de drogas.

#OBS: A função "display" imprime na tela assim como a função "print", porém de uma forma mais "arrumada"
```

```python
# Tamanho do DataFrame:
display(dados.shape)
## (rows, columns) = (linhas, colunas)

# Imprimir cabeçalho, default: primeiras 5 linhas:
display(dados.head())
```

```python
# Separar uma coluna, imprime a coluna "tratamento"
print(dados['tratamento'])

# Valores únicos da coluna:
print('\nValores únicos da tabela Tratamento: ')
print(dados['tratamento'].unique())
```

```python
# Valores únicos da coluna:
print('Valores únicos da tabela Tempo: ')
print(dados['tempo'].unique())

print('\nValores únicos da tabela Dose: ')
print(dados['dose'].unique())

print('\nValores únicos da tabela Droga: ')
print(dados['droga'].unique())

print('\nValores únicos da tabela g-0: ')
dados['g-0'].unique()
```

```python
# Contar a frequência dos valores da variável, o quanto o mesmo valor apareceu:
print(dados['tratamento'].value_counts())
```

```python
#Desafio 05: Procurar na documentação o método query(pandas)
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html?highlight=query#pandas.DataFrame.query

'''
DataFrame.query(expr, inplace=False, **kwargs)
Query the columns of a DataFrame with a boolean expression.

Parameters
expr: str
The query string to evaluate.

You can refer to variables in the environment by prefixing them with an ‘@’ character like @a + b.

You can refer to column names that are not valid Python variable names by surrounding them in backticks. Thus, column names containing spaces or punctuations (besides underscores) or starting with digits must be surrounded by backticks. (For example, a column named “Area (cm^2) would be referenced as Area (cm^2)). Column names which are Python keywords (like “list”, “for”, “import”, etc) cannot be used.

For example, if one of your columns is called a a and you want to sum it with b, your query should be `a a` + b.

New in version 0.25.0: Backtick quoting introduced.

New in version 1.0.0: Expanding functionality of backtick quoting for more than only spaces.

inplace: bool
Whether the query should modify the data in place or return a modified copy.

**kwargs
See the documentation for eval() for complete details on the keyword arguments accepted by DataFrame.query().

Returns
DataFrame or None
DataFrame resulting from the provided query expression or None if inplace=True.
'''
#Exemplo de aplicação do método query:
print('Utilizando método query:')
print(dados.query('`tempo` == 24').shape)
##Retorna o número de linhas e colunas para 'tempo' igual a 24.

#Equivalente à:
print('Não utilizando método query:')
print(dados[dados.tempo == 24].shape)
```

```python
#Desafio 06: Renomear as colunas tirando o hífen.

#Cria uma cópia do Dataset para fins de manter o original intacto:
dados_copy = dados.copy()

#Imprime o dataset (antes):
display(dados_copy.head())

#laço para percorrer os valores de 0 a 100:
for j in range(0, 100):
  dados_copy = dados_copy.rename(columns={f"g-{j}": f"g{j}", f"c-{j}": f"c{j}"})

#Imprime o dataset (depois):
display(dados_copy.head())
```

```python
# Imprime frequência e proporção entre os valores únicos da variável:
print('Frequência dos valores de dose:')
print(dados['dose'].value_counts(normalize=False))
print('\nProporção dos valores de dose:')
print(dados['dose'].value_counts(normalize=True))
```
